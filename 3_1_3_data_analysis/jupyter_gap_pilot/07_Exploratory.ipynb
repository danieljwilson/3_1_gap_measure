```r
cor(df, use="pairwise.complete.obs") %>%
    as.data.frame() %>%
    mutate(var1 = rownames(.)) %>%
    gather(var2, value, -var1) %>%
    arrange(desc(value)) %>%
    group_by(value) %>%
    filter(row_number()==1)
```

## Gender Differences

```r
df = qn_df %>% 
    select(starts_with('ib_domain_success')) %>% 
    mutate_all(as.numeric)
```

```r
names(df) <- substring(names(df), 19)
```

```r
df = cbind(df, qn_df[,'subjectID'])
```

```r
df = merge(df, demographics[, c('gender', 'ethnicity', 'subjectID')], by='subjectID')
df = merge(df, ib_measures_df[, c('domain_gap', 'subjectID')], by='subjectID')
```

```r
# remove NAs from test-retest subjects
df = df[!is.na(df$gender),]
```

```r
df$goalCount = rowSums(!is.na(df[,2:33]))
```

```r
head(df)
```

```r
df[1]
```

```r
males = subset(df, gender == 'Male')
females = subset(df, gender== 'Female')
```

### Goal Count

```r
df %>% group_by(gender) %>%
    summarise(goal_count=mean(goalCount),
              n = n()             
             )
```

```r
# make sure variance in both samples is homogenous to pass test of homoscedasticity
var.test(males$goalCount, females$goalCount)
```

```r
t.test(males$goalCount, females$goalCount, var.equal = TRUE)
```

```r
options(repr.plot.width = 20, repr.plot.height = 10)

ggplot(data=df, aes(x=goalCount, group=gender, fill=gender)) +
    geom_density(adjust=1.5, alpha=.4) +
    theme_minimal()
```

### Gap

```r
df %>% group_by(gender) %>%
    summarise(gap=mean(domain_gap))
```

```r
t.test(males$domain_gap, females$domain_gap)
```

```r
options(repr.plot.width = 20, repr.plot.height = 10)

ggplot(data=df, aes(x=domain_gap, group=gender, fill=gender)) +
    geom_density(adjust=1.5, alpha=.4) +
    theme_minimal()
```
